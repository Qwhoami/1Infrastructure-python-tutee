# if

```python
# if
# indentation 주의

gate = True
if gate:
  print('yes')
else:
  print('no')
```

```python
# 비교 연산자
# < , >, == , !=, >=, <=
if 1> 2:
  print('no')
else:
  print('yes')
```

```python
# or and not

a = [1,2,3]
b = [3,4,5]

if 2 in a and 2 in b:
  print('good')
else:
  print('bad')


if 2 not in a:
  print('yes')
else:
  print('no')

```

```python
# if문에는 조건이 들어가야함 안들어가기 원하면 pass
a = [1,2,3]
b = [3,4,5]

if 2 in a and 2 in b:
  print('good')
else:
  pass
print('end of if statements')
```

# 반복문

```python
# while True
# 조건을 만족할때 까지만 실행
a = 0
while a < 5:
  a += 1
  print(a)

print('End',a)


# VScode Debugging:
```

# For loop

```python
# for 변수 in iterable variable (list, tupel, ,,,):
a = [1,2,3]
for i in a:
  print(i)
```

```python
a = [(1,2), (2,3), (3,4)]
for (b, c) in a:
  print(b, c)

for b, c in a:
  print(b, c)
```

```python
for i in range(1, 11):
  print(i)

print('='*50)

for i in range(1, 11, 2):
  print(i)
```

```python
out = [    num for num in [1,2,3]      ]
print(out)


out = [    num for num in [1,2,3] if num % 2 != 0  ]
print(out)
```
