```python
# open(path, mode)
# w: 파일에 내용을 쓸 때, r: 파일을 읽기만 할 때, a: 파일의 마지막에 새로운 내용을 추가 할 때 (file이 계속 overlap 되기때문에 주의, log history file 만들때 사용)
rel_path = 'kjr_test.txt'
f = open(rel_path, 'w')
# f = open(rel_path, 'a') # 덮어쓰기
[f.write(f'{i+1}\n') for i in range(10)]
f.close()
```

```python
f = open(rel_path, 'r')
print(f.readline())
print(f.readline())
f.close()
# print(f.readline()) # close 이후는 읽을 수 없음, close를 하지 않으면 나도 모르게 값이 추가 될 수 있어 debugging 어려움
```

```python
f = open(rel_path, 'r')
print(f.read())
```

```python
f = open(rel_path, 'r')
print(f.readlines())
```

```python
# f.close()를 사용 안하고 싶을때
with open('test2.txt', 'w') as f:
  f.write('hello')
# f.write('test') # indentation으로 close 구분
```

```python
import sys
args =  sys.argv[1:] #0: file.py, 1~: arguments 입력
for i in args:
  print(i)

# python3 file.py 1 2 3
```

```python
with open('arguments.py', 'w') as f:
  f.write('import sys\n')
  f.write('args =  sys.argv[1:]\n')
  f.write('for i in args:\n')
  f.write('  print(i, "test")\n')
```
