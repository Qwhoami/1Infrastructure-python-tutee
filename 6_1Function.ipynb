```python
def sub(a,b): #a, b are parameter
  print('')
  # return a-b # retun을 만나면 함수가 끝나는것이므로 주의
  # return 사용 안하면 none 으로 출력

print(sub(1,3)) #1,3 are argument
print(sub(b=3, a=1))
```

```python
def many_args(*argss): # asterisk variable: positional arguments, 여러 개의 인수를 튜플로 모아 받을 수 있음
  print(type(argss))
  print(argss)

many_args(3,5,6,7,8, 5, 3)
```

```python
def dict_args(**kwargs): # double asterisk variable: keyward arguments, 여러 개의 인수를 dictionary로 모아 받을 수 있음
  print(type(kwargs))
  print(kwargs)
  print(kwargs.keys())
  print(kwargs.items())

dict_args(a=1, b=1)
```

```python
def test(a,b,c=1): # 기본값이 지정된 parameter는 뒤에 위치해야함,
  print(a, end=' start\n')
  print(b)
  print(c, end=' end\n\n')

test(1,2)
test(1,2,4)
```

```python
# gloabl & local variable

a=1 # Immutable (tuple, number, string)

def test(a):
  a = a+1 # local variable

test(a)
print(a)
```

```python
a=1

def test():
  global a
  a = a+1

test()
print(a)
```

```python
# mutable compared to immutable

a = [3,4,5] # Mutable

def test(a):
  a.append(9) # Mutable list 자체가 변함
              # 함수 종료하면서 지역변수 사라짐

print(test(a)) # 함수의 return이 없기에, 지역변수 사라지고 none 출력
print(a)
```

```python
# def add(a,b):
#   return a+b

add = lambda a,b: a+b # add라는 함수는 a,b를 매개변수로 사용 : 뒤애 return 값을 기재해 간단하게 함수 표현 가능
print(add(3,2))


# 함수의 이름을 정하지 않는 장점이 있음
test = [lambda a,b: a*b, 1,2,3] # list=[f(x), 1, 2, 3]
test = [test[0](3,7)] + test[1:]
print(test)

```
