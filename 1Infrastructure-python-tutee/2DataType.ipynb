```python
# Mutable (List, Dictionary, Set) VS Immutable (Tuple, String, Integer, Float, Bool)
# Mutable: 값을 변경할 때마다 새로운 객체를 생성해야 하므로, 잦은 변경 작업에서는 비효율
# Immutable: 변경 불가로 코드가 안전하고 이해하기 쉬움
# List와 거의 대동상의 Mutable (List, Dictionary, Set) VS Immutable (Tuple, String, Integer, Float, Bool)
# Mutable: 값을 변경할 때마다 새로운 객체를 생성해야 하므로, 잦은 변경 작업에서는 비효율
# Immutable: 변경 불가로 코드가 안전하고 이해하기 쉬움
# List와 거의 대동소이
a = [1,2,3]
a[0] = 2
print(a)

b = (1,2,3)
print(b[0])
# b[0] =2

del b[0]
```

Data type: Dictionary

```python
# 대부분의 API가 JSON 형태로 제공됨
# key는 중복되면 덮어씌어짐 주의!
a = {
      'name': 'kj',
      'age' : 34,
      'height': 180
}


print(a)

# key, value
a['personality'] = 'very nice'

print(a)

print(a['name']) # dict에 없는 값 사용시 에
print(a.get('name')) #dict에 없는 값을 호출했을때 None이 나와서 나중에 if문 사용하기 편함
print(a.get('kkk', '값이 없습니다.')) #dict에 없는 값을 호출했을때 None이 나와서 나중에 if문 사용하기 편함
print('name' not in a )

del a['age'], a['personality']
print(a)


print(a.keys())
print(a.values())
print(a.items())
```

```python
# key는 immutable만 가능
# a= {[1,2]: 1}
a= {(1,2): 1}
```

Data type: Set

```python
#중복허용X, 순서없음 --> indexing 못함
a = set('hello')
print(type(a))
print(a)

# 교집합 &, intersection, 합집합 |, union, 차집합 -, difference
a = {1,2,3}; b={2,3,4}
print(a & b)
print(a.intersection(b))
print(a | b)
print(a.union(b))
print(b-a)
print(b.difference(a))

# add 하나, update 여러개
a.add(5)
print(a)
a.update([6,7,8])
print(a)
a.remove(6)
print(a)
```
