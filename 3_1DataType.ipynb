Data type: 숫자

```python
'''
숫자
int 정수형
float 실수형
Scientific notation
2진수
8진수
16진수
'''
a =1
print(type(a))
a= 1.2
print(type(a))
a = 1e-6
print(type(a))
a = 0b011
print(a)
a = 0o110
print(a)
a = 0x10
print(a)
```
