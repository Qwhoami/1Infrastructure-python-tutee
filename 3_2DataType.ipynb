```python
my='hi'
# 왼쪽 정렬
print(f'{my:<10}!')
# 오른쪽 정렬
print(f'{my:>10}!')
# 가운데 정렬
print(f'{my:^10}!')
# 공백
print(f'{my:-^10}!')
```

```python
# Methods for string
a = 'testt'
print(a.count('t')) # return the number of string
print(a.find('e')) # return index

out = []
for test, s in enumerate(a):
  if s == 't':
    out.append(test)
print(out)

test = [i for i,s in enumerate(a) if s=='t']
print(test)


print(a.index('t'))

a = 'testt'
print(a.find('x'))
print(a.index('x'))

# find & index 모두 cpython으로 이미 compile 되어 python.exe에 포함되어있음
```

Data type: Dictionary

image.png

```python
# 대부분의 API가 JSON structure로 제공됨
# key는 중복되면 덮어씌어짐 주의!
a = {
      'name': 'kj',
      'age' : 34,
      'height': 180
}
print(a)

```

```python
# key, value
a['personality'] = 'very nice'
print(a)
```

```python
print(a['name'])
print(a.get('name'))
print(a.get('kkk', '값이 없습니다.')) #dict에 없는 값을 호출했을때 default None이 나와서 나중에 if문 사용하기 편함
print('name' not in a )

# del a['age'], a['personality']
# print(a)

print(a.keys())
print(a.values())
print(a.items())
```

```python
# key는 immutable만 가능
# a= {[1,2]: 1}
a= {(1,2): 1}
print(a)
```

Data type: Set

```python
#중복허용X, 순서없음 --> indexing 못함
a = set('hello')
print(type(a))
print(a)
# print(a[0])
```

```python
# 교집합: &, intersection/ 합집합: |, union /차집합: -, difference
a = {1,2,3}; b={2,3,4}
print('a & b -->', a & b)
print('a.intersection(b) -->', a.intersection(b))
print('a | b -->', a | b)
print('a.union(b) -->', a.union(b))
print(b-a)
print(b.difference(a))
```

```python
# add 하나, update 여러개
a.add(5)
print(a)
a.update([6,7,8])
print(a)
a.remove(6)
print(a)
```
